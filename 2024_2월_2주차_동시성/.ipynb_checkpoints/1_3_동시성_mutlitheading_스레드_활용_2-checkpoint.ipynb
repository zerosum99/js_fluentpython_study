## 1. 동기화

- 동기화는 병렬 및 동시 프로그래밍에서 여러 스레드 또는 프로세스가 데이터나 자원에 안전하게 접근하고 조작할 수 있도록 하는 메커니즘입니다. 



### 주요 동기화의 종류는 다음과 같습니다:

### Locks (락):

- 가장 기본적인 동기화 메커니즘 중 하나입니다.
- 락은 단일 스레드만 접근할 수 있는 코드 블록을 보호하고, 다른 스레드는 락이 해제될 때까지 대기합니다.
- 파이썬에서는 threading.Lock 및 threading.RLock 클래스를 사용하여 락을 구현할 수 있습니다.

### Conditions (조건):

- 여러 스레드가 특정 조건을 충족할 때까지 대기하고, 다른 스레드가 조건을 충족할 때까지 대기 중인 스레드를 깨우는 데 사용됩니다.
- 파이썬에서는 threading.Condition 클래스를 사용하여 조건을 구현할 수 있습니다.

### Semaphores (세마포어):

- 동시에 특정 자원에 접근할 수 있는 허용 가능한 스레드 또는 프로세스 수를 제어하는 동기화 도구입니다.
- 파이썬에서는 threading.Semaphore 클래스를 사용하여 세마포어를 구현할 수 있습니다.

### Events (이벤트):

- 스레드간에 상호 통신을 가능하게 하고, 하나 이상의 스레드가 이벤트 상태를 기다릴 수 있도록 해줍니다.
- 파이썬에서는 threading.Event 클래스를 사용하여 이벤트를 구현할 수 있습니다.

### Barriers (배리어):

- 여러 스레드가 동시에 특정 지점까지 도달할 때까지 대기하고, 모든 스레드가 도착한 후에 동시에 진행되도록 합니다.
- 파이썬에서는 threading.Barrier 클래스를 사용하여 배리어를 구현할 수 있습니다.

###  경합조건 (Race Condition):

- 경합조건은 둘 이상의 프로세스나 스레드가 공유된 자원에 동시에 접근하려고 할 때 발생합니다. 
- 이 때, 결과는 프로세스나 스레드의 실행 순서에 따라 달라질 수 있습니다.

- 경합조건을 피하기 위해서는 상호 배제(Mutual Exclusion)를 통해 공유 자원에 대한 동시 접근을 제한하고, 동기화 기법을 사용하여 적절한 순서로 접근하도록 조절해야 합니다. 
- 예를 들어, 락(Lock)을 사용하여 특정 부분의 코드를 임계 영역으로 만들어 여러 프로세스 또는 스레드가 동시에 실행되지 않도록 보호할 수 있습니다.

## 2.  락 동기화 

## 2-1.  상호배제 동기화 : 뮤텍스 처리 


- 스레드 간의 데이터를 안전하게 동기화하기 위해 뮤텍스(뮤텐)를 사용할 수 있습니다. 
- 뮤텍스는 락(lock)이라고도 불리며, 여러 스레드가 동시에 공유 데이터에 접근하는 것을 막기 위해 사용됩니다.

- 파이썬에서는 threading 모듈의 Lock 클래스를 사용하여 뮤텍스를 구현할 수 있습니다. 
- 뮤텍스를 사용하면 특정 코드 블록을 하나의 스레드만 실행할 수 있도록 만들 수 있습니다.

```python
Image("./image/뮤텍스.png")
```

### 공유할 데이터를 전역변수로 정의 

```python
# 공유 데이터
shared_data = 0
```

### 뮤텍스 생성  

- threading.Lock을 사용하여 modify_shared_data 함수에서 공유 데이터 shared_data를 안전하게 수정하고 있습니다. 

- with mutex: 구문을 사용하여 뮤텍스가 활성화된 동안에만 해당 코드 블록이 실행되도록 하였습니다.

```python

# 뮤텍스 생성
mutex = threading.Lock()

def modify_shared_data():
    global shared_data
    with mutex:                  # 처리되는 동안 락처리 
                                 # 뮤텍스를 사용하여 공유 데이터 안전하게 수정
        for _ in range(1000):
            shared_data += 1
```

### 스레드를 생성하고 함수에서 동일한 데이터를 갱신한다 

```python

if __name__ == "__main__":
    # 두 개의 스레드 생성
    thread1 = threading.Thread(target=modify_shared_data)
    thread2 = threading.Thread(target=modify_shared_data)

    # 각각의 스레드 시작
    thread1.start()
    thread2.start()

    # 각각의 스레드가 완료될 때까지 대기
    thread1.join()
    thread2.join()

    print("Final shared_data:", shared_data)
```

## 2.  이벤트 처리 

- 스레드를 사용하여 이벤트 루프를 처리하려면 threading 모듈과 이벤트 객체를 활용할 수 있습니다.
- 이벤트 객체는 스레드 간에 상태를 전달하고 스레드가 특정 조건을 기다리도록 하는 데 사용됩니다.


### 주의사항:

- threading.Event 객체를 사용하여 이벤트를 생성하고, is_set() 메서드로 이벤트가 설정되었는지 확인합니다.
- time.sleep 함수를 사용하여 스레드가 루프를 실행하는 동안 일시적으로 대기하도록 합니다.
- KeyboardInterrupt 예외를 처리하여 사용자가 프로그램을 중지하면 이벤트를 설정하고 스레드를 종료합니다.
- thread.join()을 사용하여 스레드가 완료될 때까지 기다립니다.

## 2-1 이벤트 처리


### Event 객체
- thread 간의 간단한 통신을 위해 사용되는 객체
 
- is_set() : 내부 플래그가 True면 그때만 True 반환
- set() : 내부 플래그를 True로 설정
- clear() : 내부 플래그를 False로 설정
- wait(timeout=None) : 내부 플래그가 True가 될때까지 blocking함. 

```python
class Test:
    def __init__(self):
        print('Test')
        self.evt = threading.Event()
        self.start()
        self.stop()
        
    def stop(self):
        print('stop')
        time.sleep(1)
        self.evt.wait()
    
    def start(self):
        print('start')
        time.sleep(1)
        self.evt.set()
        
```

## 2-2 이벤트로 동기화 처리 

```python
# 이벤트 객체 생성
event = threading.Event()
```

```python
def event_loop():
    while not event.is_set():  # 이벤트가 설정되지 않은 동안 계속 반복
        print("Event loop is running")
        time.sleep(1)
```

```python

if __name__ == "__main__":
    # 이벤트 루프를 처리할 스레드 생성
    thread = threading.Thread(target=event_loop)

    # 스레드 시작
    thread.start()

    try:
        # 일정 시간 동안 대기
        time.sleep(5)

        # 이벤트 설정
        event.set()

        # 스레드가 완료될 때까지 대기
        thread.join()
    except KeyboardInterrupt:
        # KeyboardInterrupt 예외 발생 시, 이벤트 설정 및 스레드 종료
        event.set()
        thread.join()

    print("Event loop completed")
```

## 3. 락 이해하기 

## 3-1. 교착상태 (Deadlock):

- 교착상태는 둘 이상의 프로세스가 서로의 작업이 끝나기를 기다리며 무한히 대기하는 상태를 의미합니다. 
- 이는 네 가지 조건이 동시에 만족될 때 발생합니다:

- > 상호 배제(Mutual Exclusion): 최소한 하나의 자원이 배타적으로 사용될 수 있어야 합니다.
- > 점유 대기(Hold and Wait): 최소한 하나의 자원을 점유한 상태에서 다른 자원을 기다리고 있어야 합니다.
- > 비선점(No Preemption): 다른 프로세스에 의해 현재 사용 중인 자원을 강제로 뺏을 수 없어야 합니다.
- > 순환 대기(Circular Wait): 프로세스의 집합 {P0, P1, ..., Pn}에서 P0는 P1이 점유한 자원을 대기하고, P1은 P2가 점유한 자원을 대기하며, Pn은 P0이 점유한 자원을 대기해야 합니다.


#### 교착상태를 피하기 위해서는 위 네 가지 조건 중 하나를 제거하거나, 상황에 따라 자원을 할당하는 순서를 조절하여 교착상태를 방지할 수 있습니다.

## 3-2 교착상태 발생하기

```python

# 두 개의 자원
resource1 = threading.Lock()
resource2 = threading.Lock()

# 첫 번째 프로세스
def process1():
    with resource1:
        print("Process 1 has Resource 1")
        time.sleep(1)  # 시간 지연을 추가하여 교착상태를 발생시키기 쉽게 함
        # 어떤 작업 수행
        with resource2:
            print("Process 1 has Resource 2")
            # 어떤 작업 수행

# 두 번째 프로세스
def process2():
    with resource2:
        print("Process 2 has Resource 2")
        time.sleep(1)  # 시간 지연을 추가하여 교착상태를 발생시키기 쉽게 함
        # 어떤 작업 수행
        with resource1:
            print("Process 2 has Resource 1")
            # 어떤 작업 수행

# 두 개의 프로세스를 실행
thread1 = threading.Thread(target=process1)
thread2 = threading.Thread(target=process2)

thread1.start()
thread2.start()

thread1.join()
thread2.join()

print("Execution completed")
```

## 3-2. 스레드의 lock 과 Rlock 차이 


### Lock (threading.Lock):

- Lock은 단일 스레드가 리소스를 소유하고 있을 때만 다른 스레드들이 해당 리소스에 접근하지 못하게 막는 데 사용됩니다.
- 한 스레드가 Lock을 획득하면, 다른 스레드들은 해당 Lock을 획득할 때까지 대기해야 합니다.
- 즉, Lock은 한 번에 하나의 스레드만이 임계 영역에 진입할 수 있도록 하는 도구입니다.

### RLock (threading.RLock):

- RLock은 리커런트(재진입) 락이라고도 불리며, 같은 스레드가 여러 번 RLock을 획득할 수 있습니다.
- 이는 같은 스레드가 이미 획득한 RLock을 다시 획득할 수 있게 해주며, 획득한 횟수만큼 해당 스레드는 반드시 같은 횟수로 락을 해제해야 합니다.
- 따라서 RLock을 사용하면 리소스를 여러 계층에서 사용하는 경우에 유용합니다.

```python

def example_with_lock(lock):
    with lock:
        print("Inside critical section", threading.current_thread().name)
        time.sleep(1)
        # 여기에서 다른 스레드는 대기
```

```python
# Lock 사용 예제
lock = threading.Lock()
thread1 = threading.Thread(target=example_with_lock, args=(lock,))
thread2 = threading.Thread(target=example_with_lock, args=(lock,))
thread1.start()
thread2.start()
thread1.join()
thread2.join()
```

```python
def example_with_rlock(rlock):
    with rlock:
        print("Inside critical section",threading.current_thread().name)
        time.sleep(1)
        # 같은 스레드가 다시 획득해도 OK
        with rlock:
            print("Inside nested critical section",threading.current_thread().name)
            time.sleep(1)
```

```python
# RLock 사용 예제
rlock = threading.RLock()
thread3 = threading.Thread(target=example_with_rlock, args=(rlock,))
thread4 = threading.Thread(target=example_with_rlock, args=(rlock,))
thread3.start()
thread4.start()
thread3.join()
thread4.join()
```

## 4. Condition 클래스 

- 파이썬의 threading 모듈에서 제공되는 동기화 도구 중 하나로, 스레드 간의 통신 및 조건부 실행을 가능하게 합니다. 
- Condition은 Lock 객체와 함께 사용되어 공유된 리소스에 대한 안전한 접근 및 조건부 실행을 지원합니다.



### Condition 클래스가 하는 주요 역할은 다음과 같습니다:

#### 스레드 간 통신:

- Condition은 스레드 간에 신호를 보내고 기다리는 기능을 제공합니다.
- notify() 및 notifyAll() 메서드를 사용하여 대기 중인 스레드에게 신호를 보내고, wait() 메서드를 사용하여 스레드가 신호를 기다리도록 합니다.

#### 조건부 실행:

- Condition은 특정 조건이 충족될 때까지 스레드가 대기하고, 조건이 충족되면 작업을 수행할 수 있도록 합니다.
- acquire() 및 release() 메서드를 사용하여 Condition 객체 내에서 크리티컬 섹션을 지정하고, wait() 메서드를 호출하여 조건이 충족될 때까지 스레드를 대기 상태로 만듭니다.

## 4-1 조건부 실행 동기화하기 

-  SharedResource 클래스 내의 produce 및 consume 메서드에서 Condition을 사용하여 스레드 간의 조건부 실행을 동기화합니다.

```python
class SharedResource:
    def __init__(self):
        self.resource = None
        self.condition = threading.Condition()

    def produce(self, item):
        with self.condition:
            # 생산자 스레드가 작업을 수행하기 전에 조건을 확인
            while self.resource is not None:
                self.condition.wait()  # 조건이 충족되지 않으면 대기

            # 생산자가 아이템을 생성하고 공유 리소스에 할당
            self.resource = item
            print(f"Produced: {item}")

            # 소비자 스레드에게 신호를 보내어 깨움
            self.condition.notify()

    def consume(self):
        with self.condition:
            # 소비자 스레드가 작업을 수행하기 전에 조건을 확인
            while self.resource is None:
                self.condition.wait()  # 조건이 충족되지 않으면 대기

            # 소비자가 아이템을 소비하고 공유 리소스 초기화
            item = self.resource
            self.resource = None
            print(f"Consumed: {item}")

            # 생산자 스레드에게 신호를 보내어 깨움
            self.condition.notify()

```

```python
# 공유 리소스 객체 생성
shared_resource = SharedResource()

# 생산자 스레드
def producer():
    for i in range(5):
        shared_resource.produce(i)

# 소비자 스레드
def consumer():
    for i in range(5):
        shared_resource.consume()

# 스레드 생성 및 실행
producer_thread = threading.Thread(target=producer)
consumer_thread = threading.Thread(target=consumer)

producer_thread.start()
consumer_thread.start()

producer_thread.join()
consumer_thread.join()
```

## 5. 세마포어 동기화 처리

### 세마포어(Semaphore)
- 병렬 프로그래밍에서 동기화를 위한 도구로 사용됩니다. 
- 세마포어는 특정 자원에 대한 접근을 제어하고, 공유 자원의 동기화를 달성하는 데 사용됩니다. 
- 파이썬에서는 threading 모듈이나 multiprocessing 모듈을 사용하여 세마포어를 구현할 수 있습니다.

###  threading.Semaphore를 사용

- 최대 2개의 스레드만 동시에 update_resource 메서드를 실행할 수 있도록 제한하고 있습니다. 
- 세마포어는 with 문을 사용하여 동시에 하나의 스레드만이 세마포어를 획득하고, 나머지는 대기 상태로 전환되어 실행을 기다리게 합니다. 
- 이를 통해 공유 자원에 대한 안전한 동기화가 이루어집니다.

## 5-1 세마포어로 동기화 처리 

```python
import threading

class SharedResource:
    def __init__(self):
        self.resource = 0
        self.semaphore = threading.Semaphore(value=2)  # 최대 2개의 스레드 동시 실행 허용

    def update_resource(self, value):
        with self.semaphore:
            current_thread = threading.current_thread()
            print(f"{current_thread.name} is updating resource.")
            self.resource += value
            print(f"Resource value after update: {self.resource}")
```

```python
# 공유 리소스 객체 생성
shared_resource = SharedResource()

# 스레드에서 공유 리소스를 동시에 업데이트하는 함수
def thread_function():
    for _ in range(3):
        shared_resource.update_resource(1)

# 여러 스레드 생성
thread1 = threading.Thread(target=thread_function, name="Thread-1")
thread2 = threading.Thread(target=thread_function, name="Thread-2")
thread3 = threading.Thread(target=thread_function, name="Thread-3")

# 스레드 시작
thread1.start()
thread2.start()
thread3.start()

# 스레드 대기
thread1.join()
thread2.join()
thread3.join()

print("Execution completed")
```

## 6. 배리어 처리 

### Barrier(배리어)
- 병렬 프로그래밍에서 스레드나 프로세스 간의 동기화를 달성하기 위한 도구 중 하나입니다. 
- 배리어는 지정된 개수의 스레드나 프로세스가 특정 지점에서 동시에 만날 때까지 기다리는 역할을 합니다.


## 6-1 배리어 처리 

```python
# 공유 리소스 객체 생성
shared_resource = 0

# 배리어 객체 생성 (3개의 스레드가 만날 때까지 기다림)
barrier = threading.Barrier(3)

# 스레드에서 공유 리소스를 동시에 업데이트하는 함수
def thread_function():
    global shared_resource
    for _ in range(3):
        shared_resource += 1
        barrier.wait()  # 모든 스레드가 이 지점에서 만날 때까지 기다림
        print(f"Thread {threading.current_thread().name} updated resource to {shared_resource}")

# 여러 스레드 생성
thread1 = threading.Thread(target=thread_function, name="Thread-1")
thread2 = threading.Thread(target=thread_function, name="Thread-2")
thread3 = threading.Thread(target=thread_function, name="Thread-3")

# 스레드 시작
thread1.start()
thread2.start()
thread3.start()

# 스레드 대기
thread1.join()
thread2.join()
thread3.join()

print("Execution completed")
```
