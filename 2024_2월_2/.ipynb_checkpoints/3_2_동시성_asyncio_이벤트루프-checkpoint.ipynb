## 파이썬의 이벤트 루프

- 비동기 프로그래밍 및 동시성 작업을 처리하는 핵심 메커니즘입니다. 이벤트 루프는 이벤트와 콜백 함수를 관리하고, 비동기 작업을 조정하며, 이벤트를 처리하는 역할을 합니다. 파이썬에서는 asyncio 라이브러리를 통해 이벤트 루프를 사용할 수 있습니다.

### 이벤트 루프는 다음과 같은 주요 기능을 제공합니다:

#### 이벤트 처리: 
- 이벤트 루프는 비동기적으로 발생하는 이벤트를 처리합니다. 
- 예를 들어, 네트워크 요청의 완료, 파일 읽기 완료 등의 이벤트를 처리할 수 있습니다.

#### 비동기 작업 관리: 
- 이벤트 루프는 asyncio를 사용하여 비동기 작업을 조정하고 실행합니다. 
- asyncio를 통해 비동기 작업을 생성하고, 실행하고, 종료할 수 있습니다.

#### 콜백 처리: 
- 이벤트 루프는 콜백 함수를 관리하고 실행합니다. 
- 비동기 작업이 완료되면 이벤트 루프는 해당 콜백 함수를 호출하여 작업을 처리합니다.

#### 동시성 제어: 
- 이벤트 루프는 동시성 작업을 제어하고 동기화를 처리합니다. 
- 여러 개의 작업이 동시에 실행되는 경우, 이벤트 루프는 동기화를 통해 상호작용을 조정합니다.

#### 예외 처리: 
- 이벤트 루프는 비동기 작업 및 이벤트 처리 중 발생하는 예외를 처리합니다. 
- 예외가 발생하면 이벤트 루프는 해당 예외를 적절히 처리하고 오류를 관리합니다.

## 현재 이벤트 루프 확인 

- asyncio.get_running_loop() 함수로 현재 이벤트루프 정보를 가져온다 
- is_running() 메서드는 이벤트 루프가 현재 실행 중인지 여부를 확인하고, 
- is_closed() 메서드는 이벤트 루프가 닫혀 있는지 여부를 확인합니다.

```python
def get_running_loop():
    loop = asyncio.get_running_loop()
    print("Current running event loop:", loop)
    print("Is event loop running?", loop.is_running())
    print("Is event loop closed?", loop.is_closed())

async def main():
    get_running_loop()

asyncio.run(main())
```

## 코루틴에서 전역과 지역이 존재

```python
async def coroutine_example():
    local_variable = 'Local variable in coroutine'

    def inner_function():
        nonlocal local_variable
        local_variable = 'Modified local variable in inner function'

    inner_function()
    print("Inside coroutine:", local_variable)

global_variable = 'Global variable'

async def main():
    await coroutine_example()
    print("Outside coroutine:", global_variable)


# 현재 이벤트 루프 가져오기
loop = asyncio.get_event_loop()

# 주피터 노트북에서는 이벤트 루프를 사용하기 전에 설정이 필요합니다.
# 이 설정은 주피터 노트북 환경에서 asyncio를 사용하기 위한 것입니다.
loop.run_until_complete(main())
```

## 이벤트 내의 모든 태스크 확인


- all_tasks() 함수를 사용하여 현재 이벤트 루프 내에 있는 모든 태스크를 가져와서 출력합니다. 

- asyncio.gather() 함수를 사용하여 메인 함수가 모든 태스크를 완료할 때까지 기다립니다.

```python
async def task1():
    await asyncio.sleep(1)
    print("Task 1 completed")

async def task2():
    await asyncio.sleep(2)
    print("Task 2 completed")

async def main():
    # 태스크 생성
    t1 = asyncio.create_task(task1())
    t2 = asyncio.create_task(task2())

    # 현재 이벤트 루프 내에 있는 모든 태스크 확인
    tasks = asyncio.all_tasks()
    print("Tasks in the event loop:", tasks)

    # 메인 태스크 완료 대기
    await asyncio.gather(t1, t2)

asyncio.run(main())
```

## 1-1 코루틴 등록

- asyncio는 코루틴을 사용하여 이벤트를 등록합니다. 
- 비동기 함수나 메서드를 정의하고, 이를 asyncio의 이벤트 루프에 등록합니다. 
- 코루틴은 이벤트가 발생했을 때 실행되어 비동기적으로 작업을 처리합니다.

```python
# 비동기 작업을 처리할 코루틴 함수 정의
async def my_coroutine():
    print("Coroutine is starting")
    # 비동기적으로 대기하고 작업을 수행
    await asyncio.sleep(1)
    print("Coroutine is done")

# 이벤트 루프에서 코루틴을 실행하는 함수 정의
async def main():
    # 코루틴을 이벤트 루프에 등록하여 실행
    await my_coroutine()

# asyncio 이벤트 루프 생성 및 실행
asyncio.run(main())
```

## 1-2 Future 객체 등록: 

- asyncio는 Future 객체를 사용하여 이벤트를 등록합니다.
- Future 객체는 비동기 작업의 결과를 나타내며, 이벤트 루프에 등록되어 해당 작업의 완료를 기다립니다.
- 비동기 작업이 완료되면 Future 객체는 이벤트 루프에 등록된 콜백 함수를 호출하여 결과를 처리합니다.

```python
asyncio.ensure_future
```

```python
help(asyncio.ensure_future)
```

```python
# 비동기 작업을 처리할 함수 정의
async def my_task():
    print("Task is starting")
    # 비동기적으로 대기하고 작업을 수행
    await asyncio.sleep(1)
    print("Task is done")
    return "Task Result"

# 이벤트 루프에서 Future 객체를 사용하여 작업 등록
async def main():
    # 비동기 작업을 위한 Future 객체 생성
    future = asyncio.ensure_future(my_task())

    # 작업이 완료될 때까지 기다림
    await future

    # 작업 결과 확인
    print("Task result:", future.result())

# asyncio 이벤트 루프 생성 및 실행
asyncio.run(main())
```

## 1-3  이벤트 핸들러 등록: 
- asyncio는 이벤트 핸들러를 사용하여 이벤트를 등록합니다. 
- 이벤트 핸들러는 asyncio의 이벤트 루프에 등록되어 특정 이벤트가 발생했을 때 실행됩니다. 
- 이벤트 핸들러는 특정 조건이 충족되었을 때 호출되어 비동기적으로 작업을 처리합니다.

```python
# 이벤트를 처리할 핸들러 함수
async def handle_event():
    print("Handling event...")
    # 비동기적으로 대기하고 작업을 수행
    await asyncio.sleep(1)
    print("Event handled")

# 이벤트 루프에 핸들러 등록
async def main():
    # 이벤트 루프에 핸들러 등록
    loop = asyncio.get_event_loop()
    loop.create_task(handle_event())

    # 이벤트 루프 실행
    await asyncio.sleep(2)  # 이벤트 핸들러가 완료되기를 기다림

# asyncio 이벤트 루프 생성 및 실행
asyncio.run(main())
```

## 1-4 이벤트 생성 및 등록: 
- asyncio는 사용자가 직접 이벤트를 생성하고 등록할 수도 있습니다. 
- 사용자는 asyncio의 이벤트 루프에 이벤트를 등록하여 비동기적으로 처리할 수 있습니다. 
- 이벤트는 사용자가 정의한 조건에 따라 생성되고 등록될 수 있습니다.

```python
asyncio.Event
```

```python
asyncio.create_task
```

```python
# 이벤트를 처리할 코루틴 함수
async def handle_event(event):
    print("Handling event...")
    # 비동기적으로 대기하고 작업을 수행
    await asyncio.sleep(1)
    print("Event handled")

# 메인 함수
async def main():
    # 새로운 이벤트 생성
    event = asyncio.Event()

    # 이벤트를 처리할 핸들러 등록
    handler_task = asyncio.create_task(handle_event(event))

    # 이벤트 발생
    print("Event is set")
    event.set()

    # 핸들러 완료 대기
    await handler_task

# asyncio 이벤트 루프 생성 및 실행
asyncio.run(main())
```

### 이벤트 핸들러 처리 

```python
# 이벤트를 처리할 핸들러 함수
async def event_handler(event):
    print("Event handler: Waiting for event...")
    await event.wait()  # 이벤트가 설정될 때까지 대기
    print("Event handler: Event is set!")

# 메인 함수
async def main():
    # 이벤트 객체 생성
    event = asyncio.Event()

    # 이벤트 핸들러 등록
    event_handler_task = asyncio.create_task(event_handler(event))

    # 이벤트 설정
    await asyncio.sleep(2)  # 어떤 작업을 수행하고 나중에 이벤트 설정
    event.set()  # 이벤트 설정

    # 이벤트 핸들러 완료 대기
    await event_handler_task

# asyncio 이벤트 루프 실행
asyncio.run(main())
```
